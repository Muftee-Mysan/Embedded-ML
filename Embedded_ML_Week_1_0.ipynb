# examples: [(-2, -8), (-1, -5), (0, -2), (1, 1), (2, 4), (3, 7)]

```python
X = [-2, -1, 0, 1, 2, 3]  #  Features aka input
```

```python
Y = [-8, -5, -2, 1, 4, 7]  # Labels aka output
```

# By looking at the numbers, we can say that the,

# Relationship is Y = 3X - 2

```python
plt.plot(X,Y)
```

```python
def f(x):
  return 3*x - 2

y = f(x = 10)
print('y = ', y)
```

## Simplest Possible Neural Network

Dense is a layer of neurons.
units is the number of neurons in the layer.
input_shape is the number of inputs for the neurons

```python
model = keras.Sequential([
    keras.layers.Dense(units = 1, input_shape = [1])
    ])
```

This creates a Sequential model. A Sequential model is appropriate for a plain stack of layers where each layer has exactly one input tensor and one output tensor.

This line adds a Dense layer to the model.


*   Dense Layer: This is a fully connected layer. Each neuron in the layer receives input from all neurons of the previous layer.

*   units=1: This specifies that the layer will have 1 neuron.


*   input_shape=[1]: This defines the shape of the input data. Here, it expects a single feature (one-dimensional input).





Practical Example

This simple model is often used for tasks like linear regression, where there is a direct relationship between the input and output.

```python
model.summary()
```

parameters are weight and bias


optimizer is the underligned argument that does gradient descent.

check keras optimizers sgd documentatiom

```python
model.compile(optimizer = keras.optimizers.SGD(learning_rate = 0.01), loss = 'mean_squared_error')
```

optimizer=keras.optimizers.SGD(learning_rate=0.01):

SGD (Stochastic Gradient Descent):

 This is a common optimization algorithm used in machine learning. It updates the model parameters iteratively to minimize the loss function.

learning_rate=0.01:

 This sets the learning rate for the optimizer. The learning rate controls how much to change the model parameters in response to the gradient of the loss function.
loss='mean_squared_error':


Mean Squared Error (MSE):

This is a loss function often used for regression tasks. It calculates the average of the squared differences between the actual and predicted values.

# Before Training

```python
print(model.predict([10,0]))
```

```python
w, b = model.get_weights()
print(f'Before Training... \n weight = {w} \n bias = {b}')
```

```python
10.0 * w + b
```

epochs is the number of iterations

```python
hist = model.fit(X, Y, epochs = 1000)
```

```python
plt.title('Training ' + 'loss')
plt.plot(hist.history['loss'])
plt.show()
```
